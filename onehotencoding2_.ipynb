```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset/carprices (3).csv")
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis="columns")
final
```

```python
x = final.drop("Sell Price($)",axis="columns")
x
```

```python
y = final["Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
# price of Mercedez Benz that is 4 yr old with milege 45000
model.predict([[45000,4,0,0]])
```

```python
# price of bmw x5 that is 7 yr old with milege 86000

model.predict([[86000,7,0,1]])
```
