```python
# categorical variable and one hot encoding
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset/homeprices (2).csv")
df
```

```python
# using pandas to crate dummy variable
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(["town"],axis="columns")
final
```

Dummy Variable Trap When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia-0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you heed to drop one column

## **NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you**

```python
final = final.drop(["west windsor"],axis="columns")
final
```

```python
x = final.drop(["price"],axis="columns")
x
```

```python
y= final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict(x) #2600 sqrt ft home in new jersey
```

```python
model.predict([[3400,0,0]]) #3400 sqr ft home in west windsor
```

```python
model.predict([[2800,0,1]]) #2800 sqr ft home in robbinsville
```

```python
# using sklearn OneHotEncoder
# first step is to use label encoder to convert town names into number
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[["town","area"]].values
x
```

```python
y = dfle.price.values
y
```

```python
# now use one hot encoder to create dummy variables for each of the town

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("town",OneHotEncoder(),[0])], remainder = "passthrough")
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:]
x
```

```python
model.fit(x,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 srr ft home in robbinville
```
