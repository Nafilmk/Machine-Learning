```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset/HR_comma_sep (2).csv")
df.head()
```

```python
df.drop(["Work_accident","left","promotion_last_5years"],axis="columns", inplace=True)
df.head()
```

```python
x = df.drop("salary",axis="columns")
```

```python
y = df.salary.map({'low': 1, 'medium': 2, 'high': 2})
```

```python
from sklearn.preprocessing import LabelEncoder
le_department = LabelEncoder()
```

```python
x["Department_n"] = le_department.fit_transform(x["Department"])
x.head()
```

```python
z = x.drop("Department",axis="columns")
z.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(z,y,test_size=0.2)
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```
